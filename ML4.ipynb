```python
df = pd.read_csv('/content/Toyota.csv')
df.head(10)
```

```python
df['Doors'].value_counts()
```

```python
df['HP'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['FuelType']= lb.fit_transform(df['FuelType'])
df['MetColor']= lb.fit_transform(df['MetColor'])
df['Automatic']= lb.fit_transform(df['Automatic'])
df['Doors']= lb.fit_transform(df['Doors'])
df['Age'] = lb.fit_transform(df['Age'])
df['KM'] = lb.fit_transform(df['KM'])
df['HP'] = lb.fit_transform(df['HP'])
df['CC'] = lb.fit_transform(df['CC'])
df['Price'] = lb.fit_transform(df['Price'])
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='CC')
y = df['CC']
```

```python
x_test , x_train , y_test , y_train = train_test_split(x , y , test_size = 0.2 , random_state = 1)
```

```python
np.round(x_test.describe(),2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
```

```python
x_test_mn = mn.fit_transform(x_test)
```

```python
x_test_new = pd.DataFrame(x_test_mn , columns=x_test.columns)
```

```python
np.round(x_test_new.describe(),2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
x_test_lr = lr.fit(x_test,y_test)
```

```python
x_test_new = lr.predict(x_test)
```

```python
print("Slope (coef):", lr.coef_)
print("Intercept:", lr.intercept_)

# Prediction
# print("Salary for 6 years exp:", lr.predict([[6]]))
```
