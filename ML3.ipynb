```python
df = pd.read_csv('/content/click -.csv')
df.head(3)
```

```python
df = df.drop(columns=['Ad Topic Line','City','Country','Timestamp'])
```

```python
df.head(3)
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
```

```python
df.head(2)
```

```python
df['Area Income'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['Gender'])
y = df['Gender']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc,columns=x.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn,columns=x.columns)
```

```python
np.round(x_train_new.describe(),2)
```
